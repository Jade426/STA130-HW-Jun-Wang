#Question 1
A Classification Decision Tree is a machine learning model designed to solve classification tasks by categorizing observations into predefined groups based on their features. It works by creating a tree-like structure of decision rules learned from the data, guiding how items are classified step by step.

a) 
-Medical Diagnosis: Identifying if a patient has a particular disease by analyzing symptoms and medical test results, classifying as "diabetes" or "no diabetes".  
-Fraud Detection: Assessing whether a transaction is fraudulent by examining factors like the amount, location, and timing of the transaction.  
-Email Classification: Sorting emails into categories such as "spam" or "not spam" based on their content and metadata.  

b)
A Classification Decision Tree works by splitting data into branches based on feature values, using a sequence of yes/no decisions to classify items until reaching a leaf node that assigns a label, like "spam" or "not spam." It evaluates splits using metrics like Gini Impurity or Information Gain to ensure the categories are well-separated. On the other hand, Multiple Linear Regression focuses on predicting continuous outcomes, such as house prices or sales revenue, by establishing a linear equation that represents the relationship between features and the target variable. This equation combines weighted sums of the features and aims to minimize prediction errors. While decision trees rely on threshold-based rules for classification, regression uses mathematical models to deliver precise, continuous results.

#Question 2
-Accuracy: Accuracy is best suited for scenarios where false positives and false negatives carry similar consequences. For instance, during routine health checkups aimed at diagnosing common conditions like the flu or anemia, the primary goal is to achieve a high proportion of correct diagnoses, whether identifying the illness (true positives) or confirming its absence (true negatives). In these cases, accuracy offers a balanced measure of the model's overall effectiveness.

-Sensitivity
Sensitivity is crucial in situations where missing true positives (false negatives) can lead to serious consequences. For example, in cancer screening, failing to identify a true positive could delay treatment, potentially worsening the patient's prognosis. By prioritizing sensitivity, the focus shifts to detecting as many actual cases as possible, even if it increases the likelihood of false positives.

-Specificity
Specificity is essential in scenarios where correctly identifying true negatives (avoiding false positives) is a priority. For instance, in blood donation screenings for diseases like HIV, a false positive might lead to discarding viable blood and causing unnecessary distress to healthy donors. By maximizing specificity, the focus is on reliably excluding disease-free cases, ensuring efficient and accurate results.

-Precision
Precision is critical in situations where false positives carry significant costs or cause disruption. For instance, in fraud detection, incorrectly flagging a legitimate transaction can lead to customer frustration and a loss of trust. By prioritizing precision, the system ensures that flagged transactions are more likely to represent actual fraud, reducing false alarms and minimizing inconvenience for users.

#Question 9
In linear regression, coefficients provide a clear and interpretable measure of how the dependent variable changes with a one-unit increase in a predictor, assuming all other variables remain constant. These coefficients convey both the magnitude and the direction of the effect, making it easy to understand the precise relationship between predictors and the target variable.

In contrast, feature importances in decision trees reflect the relative contribution of each feature to the model's predictive power. They are calculated based on how much each feature reduces impurity across all splits where it is used. However, feature importances do not reveal the direction (positive or negative) or the exact magnitude of the feature's impact on the target variable.

Thus, while linear regression coefficients enable direct and quantitative interpretation of predictor effects, decision tree feature importances provide a comparative ranking of feature relevance without offering explicit information about the size or direction of their effects.
